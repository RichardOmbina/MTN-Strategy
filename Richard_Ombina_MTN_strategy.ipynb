```
# Importing libraries to be used
import pandas as pd
import numpy as np
```

```
#Loading datasets
cells_df = pd.read_csv('cells_geo.csv', sep=";", encoding='utf-8')

column_names = ['product', 'bill',	'datetime','cell_on_site','caller_anonymized', 'recipient_anonymized', 'country_a','country_b','cell_id', 'site_id']
#T1,T2,T3 are telcome datasets
T1 = pd.read_csv('Telcom_dataset.csv', encoding='utf-8', names=column_names, header=0)
T2 = pd.read_csv('Telcom_dataset2.csv', encoding='utf-8',names = column_names, header=0)
T3 = pd.read_csv('Telcom_dataset3.csv', encoding='utf-8', names = column_names, header=0)
```

```
#confirming the dataset is loaded
cells_df
```

```
# Droping unnamed column
cells_df.drop('Unnamed: 0', axis=1, inplace=True)
```

```
#confirming column is droped
cells_df.head(10)
```

```
# droping irrelevant columns
droped = ['cell_on_site','country_a','country_b']
T1.drop(droped, axis=1,inplace=True)
T2.drop(droped, axis=1,inplace=True)
T3.drop(droped, axis=1,inplace=True)
```

```
T1.head(10)
```

```
# Check for missing values
cells_df.isnull().sum()
```

```
# Droping duplicates
cells_df.drop_duplicates(inplace=True)
T1.drop_duplicates(inplace=True)
T2.drop_duplicates(inplace=True)
T3.drop_duplicates(inplace=True)
```

```
# Joining dataframe T1,T2,T3,
joined_df = pd.concat([T1,T2,T3], axis=0, join='outer',ignore_index=True)
joined_df.head(10)
```

```
merged_df = pd.merge(joined_df, cells_df, how='inner', left_on=['site_id'], right_on=['SITE_CODE'])
merged_df.head()
```

```
merged_df.shape
```

```
#drop columns 
to_drop = ['recipient_anonymized','cell_id','CELL_ID','SITE_CODE']
merged_df.drop(to_drop, axis=1, inplace=True)
```

```
#defining high traffic regions
high_traffic_regions = ['LAGUNE','LACS','BAS SASSANDRA','HAUT SASSANDRA','VALLEE DU BANDAMA','SUD BANDAMA']

region_df = merged_df[merged_df['REGION'].isin(high_traffic_regions)]
cities_regions = region_df['VILLES'].value_counts()
high_cities_regions = cities_regions[cities_regions > 500]
high_impact_cities = high_cities_regions | high_cities_regions
cities_df = merged_df[merged_df['VILLES'].isin(high_impact_cities.index)]
services = cities_df['product'].value_counts(normalize=True)
services
```

**SMS is the most used service**

```
#product bringing more money
profits = cities_df.groupby(['product'])['bill'].sum()
profits
```
